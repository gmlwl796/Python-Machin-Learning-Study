## KNN 최근접이웃갯수를 1~10까지 자동으로 입력하는 방식으로 분류예측을 실시하시오. 

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
training_accuracy = []
test_accuracy = []
```

```python
# 최근접 이웃갯수 리스트객체 준비
neighbors_settings = range(1, 11)
list(neighbors_settings)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
# 최근접 이웃갯수에 따른 모델피팅
for k in neighbors_settings:
    # knn객체생성
    knn = KNeighborsClassifier(n_neighbors = k)
    
    # knn객체에 훈련데이터 피팅
    knn.fit(X_train, y_train)
    
    # 훈련데이터 피팅결과 정확도 저장
    training_accuracy.append(knn.score(X_train, y_train)) # 리스트명.append => 리스트 맨 뒤에 값 하나 추가
    
    # 테스트테이터와의 검증결과 정확도 저장
    test_accuracy.append(knn.score(X_test, y_test))
    
    print("-------k의 갯수가 %d일 때,-----------------------------------------" % k )
    y_pred = knn.predict(X_test)
    print("\nClassification_report :")
    print(classification_report(y_test, y_pred))
    print("\nConfusion_matrix :")
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    print("\n")
```

## k갯수 지정에 따른 훈련셋과 테스트셋 정확도 성능을 <br> 비교테이블 및 러닝커브 곡선으로 제시하시오.

```python
# k갯수 지정에 따른 훈련셋과 테스트셋 정확도 성능비교
import pandas as pd

training_accuracy_sr = pd.Series(training_accuracy)
test_accuracy_sr = pd.Series(test_accuracy)

model_perf = pd.DataFrame({'K지정 갯수':list(neighbors_settings),
                           '훈련셋 정확도': training_accuracy,
                           '테스트셋 정확도': test_accuracy,
                           '정확도 차이': test_accuracy_sr - training_accuracy_sr})
model_perf
```

```python
round(model_perf, 3)
```

```python
# k갯수 지정에 따른 훈련셋과 테스트셋 모델피팅 결과 시각화
import mychar
```

```python
# 최근접 이웃개수에에 따른 모델피팅 결과 시각화
plt.plot(neighbors_settings, training_accuracy, label = '훈련셋 정확도')
plt.plot(neighbors_settings, test_accuracy, label = '테스트셋 정확도')

plt.xlabel('k_neighbors 지정갯수')
plt.ylabel('정확도')
plt.legend()
```

- 이웃의 갯수가 적을수록 모델이 복잡해지며, 훈련셋 정확도는 높게 학습되지만 테스트셋은 상대적으로 부정확한 과대적합 상태임
- 이웃의 갯수가 늘어나면서 모델이 단순해지며, 훈련셋 정확도는 낮아지지지만 테스트셋의 정확도는 향상됨 일반화 상태임
- 이웃의 갯수가 많이지만 모델이 너무 단순해져, 훈련셋 정확도도 낮고 테스트셋 정확도고 낮은 과소적합상태임

- 이웃의 갯수가 1개인 경우에는, 데이터가 30개라고 했을 때, 30개의 모든 데이터가 하나의 이웃에 맞춰야하기 때문에 복잡도가 높아지는 것
- 테스트데이터는 학습한 적이 없는 데이터니까 오분류가 상대적으로 많이 되는 것, 복잡하게 맞출 필요가 없다고 해야하나?  그러니까 k값이 증가할수록 복잡도가 더 많아짐.


- 결과 >> 이웃의 갯수가 7개일 때, 훈련 셋과 테스트셋의 정확도가 가장 적절하게 잘 맞는다고 할 수 있다.

## 로지스틱회귀모형의 규제파라미터를 5개 정도 자동으로 입력하는 방식으로 <br> 분류예측을 실시하시오.

```python
c_parameter = [0.001, 0.1, 1, 10, 100]
training_accuracy2 = []
test_accuracy2 = []
```

```python
# 관련 패키지-모듈-메서드(알고리즘) 로딩
from sklearn.linear_model import LogisticRegression
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

# for문을 이용해 C 파라미터값과 플로팅될 marker의 모양을 준비
for C, marker in zip([0.001, 0.1, 1, 10, 100], ['o', '^', 'v', 's', '-']):
    
    # panalty='l1'을 통해 L1규제(벌점) 방식으로 회귀계수를 추정함
    logreg = LogisticRegression(solver = 'liblinear', C = C, max_iter = 1000).fit(X_train, y_train)
    print("C={:.3f} l1인  로지스틱 회귀의 훈련 정확도: {:.2f}".format(C, logreg.score(X_train, y_train)))
    print("C={:.3f} l1인 로지스틱 회귀의 테스트 정확도: {:.2f}".format(C, logreg.score(X_test, y_test)))
    print("C={:.3f} l1인 로지스틱 회귀의 사용피처 갯수: {}".format(C, np.sum(logreg.coef_ != 0)))
    y_pred = logreg.predict(X_test)
    print("\nClassification_report :")
    print(classification_report(y_test, y_pred))
    print("\nConfusion_matrix :")
    cm = confusion_matrix(y_test, y_pred)
    print(cm)
    print("-----------------------------------------------------------------")
    # 훈련데이터 피팅결과 정확도 저장
    training_accuracy2.append(logreg.score(X_train, y_train)) # 리스트명.append => 리스트 맨 뒤에 값 하나 추가
    # 테스트테이터와의 검증결과 정확도 저장
    test_accuracy2.append(logreg.score(X_test, y_test))

    print()    
    plt.plot(logreg.coef_.T, marker, label="C={:.3f}".format(C))
    
# 그래프 축 관련 세부묘사
plt.xticks(range(X.shape[1]), X.columns, rotation=90)
xlims = plt.xlim()
plt.hlines(0, xlims[0], xlims[1])
plt.xlim(xlims)
plt.xlabel("피처")
plt.ylabel("계수 크기")

plt.ylim(-5, 5)
plt.legend(loc=3)
```


- 기본적으로 로지스틱회귀모형은 L2규제(벌점)을 사용하는 모델로 훈련모형의 피팅결과 투입한 피처:예측(독립)변수의 갯수가 그대로 유지됨
- C 파라미터를 적절히 사용해 회귀계수의 과다추정을 방지하기 위해 조율하는데, 어느 회귀계수라도 0으로 만들어 피처자체를 탈락시키지 않음

- penalty="l1"라는 파라미터를 추가하면 L1규제(벌점) 방식으로 모델을 피팅하게 되어 중요한 피처위주로 남기며, 중요하지 않은 피처의 회귀계수는 0으로 만들어 탈락시키는 피처선택(feature selection) 효과를 누릴 수 있음
