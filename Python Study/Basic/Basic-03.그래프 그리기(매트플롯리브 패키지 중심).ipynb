# 다변량 데이터로 그래프 그리기 ─ 산점도 그래프

```python
# 결제금액(total_bill)과 팁금액(tip) 관계 산점도에
# 제3의 변수인 성별(sex)이나 테이블당 인원수(size) 변수까지 반영한 다변량 그래프

# matplot에서 색상표시를 데이터에 존재하는 숫자를 읽어서 자동으로 구현해줌
# 데이터에 문자열로 있는 경우 이를 정수숫자로 변환해 색상 인풋값으로 사용하고자 함 

# 성별변수컬럼의 문자열 레이블을 숫자로 변환해주는 사용자정의함수
def recode_sex(sex):
    if sex == 'Female':
        return 0 
    else:
        return 1
```

```python
# 성별 레이블을 숫자로 변환해 파생변수로 만들어 기존데이터셋에 추가함
tips['sex_color'] = tips['sex'].apply(recode_sex)
tips.head()
```

```python
# 전체그래프가 위치할 기본프레임 설정
scatter_plot = plt.figure() 

# 기본프레임에 그래프 격자로 서브플롯영역을 설정
axes1 = scatter_plot.add_subplot(1, 1, 1)

# 특정 서브플롯에 대해 세부옵션으로 구현
# - 기존 결제금액(total_bill)과 팁금액(tip) 관계 산점도에
#   제3의 변수인 테이블당 인원수(size)변수까지 반영한 그래프
axes1.scatter(
    x = tips['total_bill'], 
    y = tips['tip'],    
    s = tips['size'] * 10, # s: 점의 크기(size)   
    c = tips['sex_color'], # c: 점의 색상(color)
    alpha = 0.5) # 점의 투명도

axes1.set_title('Total Bill vs Tip Colored by Sex and Sized by Size') 
axes1.set_xlabel('Total Bill') 
axes1.set_ylabel('Tip') 
```

```python
fig = plt.figure() 
seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks'] 
for idx, style in enumerate(seaborn_styles):
    plot_position = idx + 1
    with sns.axes_style(style):
        ax = fig.add_subplot(2, 3, plot_position)
        violin = sns.violinplot(x='time', y='total_bill', data=tips, ax=ax)
        violin.set_title(style) 
        
fig.tight_layout() 
```
