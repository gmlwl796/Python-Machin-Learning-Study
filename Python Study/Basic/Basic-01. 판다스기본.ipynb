## 레코드(행)과 변수컬럼(열) 지정으로 데이터 추출하기

```python
# 데이터프레임에 속한 변수컬럼명 조회
df.columns
```

### * loc()메서드로 [ [행], [열] ] 지정 --> [열] 지정시 컬럼명으로만 입력가능

```python
# 행 지정
df_sub = df.loc[[3, 7], :]
print(type(df_sub))
df_sub.head()
```

```python
# 열 지정
df_sub = df.loc[:, ['year', 'pop']]
print(type(df_sub))
df_sub.head()
```

```python
# 행, 열 지정
df_sub = df.loc[[3, 7], ['year', 'pop']]
print(type(df_sub))
df_sub
```

```python
# 행, 열 지정
df_sub = df.loc[3:10, ['year', 'pop']]
print(type(df_sub))
df_sub
```

### * iloc()메서드로 [ [행], [열] ] 지정 --> [열] 지정시 컬럼번호로만 입력가능

```python
# 행 지정
df_sub = df.iloc[[3, 7], :]
print(type(df_sub))
df_sub.head()
```

```python
# 열 지정
df_sub = df.iloc[:, [2, 5]]
print(type(df_sub))
df_sub.head()
```

```python
# 행, 열 지정
df_sub = df.iloc[[3, 7], [2, 5]]
print(type(df_sub))
df_sub
```

```python
# 행, 열 지정
df_sub = df.iloc[3:7, [2, 5]]
print(type(df_sub))
df_sub
```

```python
# 행, 열 지정
df_sub = df.iloc[3:7, [2, 4, -1]]
print(type(df_sub))
df_sub
```

```python
# 행, 열 지정
df_sub = df.iloc[3:7, 2:5]
print(type(df_sub))
df_sub
```

### * iloc()메서드로 [ [행], [열] ] 지정 --> [열] 지정시 range()로 입력가능

```python
print(range(5))
print(type(range(5)))
# range() 함수는 정해진 범위내의 정수를 만들어 내는 제네레이터(generator)임
# 실제 [0, 1, 2, 3, 4]라는 정수가 들어 있는 것은 아님
# 이를 정수 데이터로 만들어 내기 위해서는 list()함수에 넣어주어야 함
```

```python
print(list(range(5)))
print(type(list(range(5))))
```

```python
# 행, 열 지정
df_sub = df.iloc[3:7, list(range(5))]
print(type(df_sub))
df_sub
```

```python
# 행, 열 지정
df_sub = df.iloc[3:7, list(range(2, 5))]
print(type(df_sub))
df_sub
```

# 그룹별 집계

```python
df.head()
```

### 집계기준변수: year(연도), 집계결과변수: lifeExp(기대수명), 집계기술함수: mean(평균)

```python
# 집계기준변수 설정
df.groupby('year')
```

```python
# 집계기준변수와 집계결과변수 설정
df.groupby('year')['lifeExp'] 
```

```python
# 집계기준변수와 집계결과변수, 집계합수 설정: 시리즈형식으로 집계
df.groupby('year')['lifeExp'].mean()
```

```python
# 집계기준변수와 집계결과변수, 집계합수 설정: 데이터프레임 형식으로 집계
df.groupby('year')[['lifeExp']].mean()
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
result = df.groupby('year')[['lifeExp']].mean()
result.plot()
```
