#### 반응(종속)변수 클래스별 빈도수

```python
# 반응(종속)변수 클래스별 빈도수
import numpy as np

print(cancer.target_names) # 반응(종속)변수 클래스 유형
print(np.bincount(cancer.target)) # 반응(종속)변수 클래스 빈도수

cancer_bincount = {k: v for k, v in zip(cancer.target_names, np.bincount(cancer.target))}
cancer_bincount
# - 클래스 유형별 빈도수 같이 출력
```

```python
print(type(cancer_bincount))
print(cancer_bincount.keys())
print(cancer_bincount.values())
print(cancer_bincount.items())
```

#### 반응(종속)변수 클래스별 비율
* 클래스가 불균형 상태이므로 데이터분할시 이 비율이 유지되도록 stratify = cancer.target으로 설정하는 것이 필요함 => 친화형추출
* 훈련데이터이든 테스트데이터이든 이 반응(종속)변수의 클래스 비율이 같도록 샘플링이 되도록 하는 옵션임
* 현재는 종속변수와 독립변수의 갯수가 동일하지 않기 때문에, 클래스의 비율이라도 훈련셋과 테스트셋이 같도록하기 위해 stratify를 사용
* train_test_split을 stratify 옵션 없이 사용하면 비율을 그대로 뽑아내지 못함. => 과다추청 / 과소추정이 될 가능성이 높아서 startify사용

```python
for k, v in cancer_bincount.items(): 
    pct = v / sum(cancer_bincount.values()) 
    print("%s %0.3f" % (k, pct)) 
```

## (1) 데이터분할 내용을 1개 객체에 요소별로 할당하는 방식

```python
# 데이터분할 메서드 로딩
from sklearn.model_selection import train_test_split
# - sklearn 패키지(라이브러리)의 model_selection 모듈 중에서
#   train_test_split()이라는 메서드(함수)를 메모로리로 로딩함
```

#### train_test_split(data, data2, test_size, train_size, stratify, random_state)
* data: 독립 변수 데이터 배열 또는 pandas 데이터프레임
* data2: 종속 변수 데이터. data 인수에 종속 변수 데이터가 같이 있으면 생략가능함
* test_size: 검증용 데이터 개수. 1보다 작은 실수이면 비율을 나타냄
* train_size: 학습용 데이터의 개수. 1보다 작은 실수이면 비율을 나타냄. test_size와 train_size 중 하나만 있어도 됨
* random_state: 무작위 샘플링 재현성을 위한 난수 시드
* stratify: 반응(결과)변수가 클래스 불균형일 때, 훈련셋과 테스트셋의 클래스도 일종의 층화형 샘플링 되도록 설정

```python
# 데이터분할 메서드를 활용해 데이터분할 실시
cancer_split = train_test_split(cancer['data'], cancer['target'], 
                                stratify = cancer['target'], random_state = 0)
# - 분할된 데이터셋을 하나의 객체에 한꺼번에 저장해 놓음
# - 클래스가 불균형 상태이므로 데이터분할시 이 비율이 유지되도록 stratify = cancer.target으로 설정하는 것이 필요함
# - 훈련데이터이든 테스트데이터이든 이 반응(종속)변수의 클래스 비율이 같도록 샘플링이 되도록 하는 옵션임

print(type(cancer_split)) 
# - 객체유형은 리스트임

print(len(cancer_split)) 
# - 분할된 리스트객체에 4개 요소로 구성됨

cancer_split
# - 분할된 내용은 리스트객체이며, 4가지 요소별로 
# 배열형식으로 데이터분할이 되었음을 알 수 있음 
```

```python
# 4개 리스트 요소를 세부 데이터셋으로 할당
X_train = cancer_split[0] # 예측(독립)변수의 훈련용 데이터셋
X_test = cancer_split[1] # 예측(독립)변수의 테스트용 데이터셋

y_train = cancer_split[2] # 반응(종속)변수의 훈련용 데이터셋
y_test = cancer_split[3] # 반응(종속)변수의 테스트용 데이터셋
```

## (2) 데이터분할 내용을 여러개 객체에 나누어 각각 할당하는 방식

```python
# 데이터분할 메서드 로딩
from sklearn.model_selection import train_test_split
# - sklearn 패키지(라이브러리)의 model_selection 모듈 중에서
#   train_test_split()이라는 메서드(함수)를 메모로리로 로딩함
```

#### train_test_split(data, data2, test_size, train_size, stratify, random_state)
* data: 독립 변수 데이터 배열 또는 pandas 데이터프레임
* data2: 종속 변수 데이터. data 인수에 종속 변수 데이터가 같이 있으면 생략가능함
* test_size: 검증용 데이터 개수. 1보다 작은 실수이면 비율을 나타냄
* train_size: 학습용 데이터의 개수. 1보다 작은 실수이면 비율을 나타냄. test_size와 train_size 중 하나만 있어도 됨
* random_state: 무작위 샘플링 재현성을 위한 난수 시드
* stratify: 반응(결과)변수가 클래스 불균형일 때, 훈련셋과 테스트셋의 클래스도 일종의 층화형 샘플링 되도록 설정

* 예측(입력, 투입, 설명, 독립)변수는 여러 개로 구성된 2차원 배열(행렬)이므로 대문자 X로 보통 표현함
* 반응(결과, 기대, 종속)변수는 한 개로 구성된 1차원 배열(벡터)이므로 소문자 y로 보통 표현함

```python
# 데이터분할 실시
X_train, X_test, y_train, y_test = train_test_split(
    cancer['data'], cancer['target'], stratify = cancer.target, random_state = 66)
# - 분할된 데이터셋을 각각의 객체에 별도로 저장해 놓음
# - 클래스가 불균형 상태이므로 데이터분할시 이 비율이 유지되도록 stratify = cancer.target으로 설정하는 것이 필요함
# - 훈련데이터이든 테스트데이터이든 이 반응(종속)변수의 클래스 비율이 같도록 샘플링이 되도록 하는 옵션임

# - train_test_split() 메서드는 Numpy 배열형식의 4개 반환값을 생성함
# - 기본적으로 train은 75%, test는 25% 데이터셋을 할당받는데,
#   test_size 매개변수를 통해 테스트셋의 비율을 지정할 수 있음
# - random.state 매개변수를 통해 샘플의 재현성을 확보함
```

#### n_neighbors의 갯수를 조절해 모델의 복잡도와 정확도를 조정함

* 이웃의 갯수가 많을수록: 결정경계가 완만해짐, 단순한 모델, 과소적합<br> - 훈련모델이 너무 단순해 자체정확도가 낮은 상태라 테스트데이터의 정확도도 낮아짐

```python
# sklearn 패키지의 datasets 모듈에 속한 cancer 메서드(함수)를 메모리로 로딩
from sklearn.datasets import load_breast_cancer

# load_cancer()메서드를 사용해 cancer 데이터셋 바로 로딩
cancer = load_breast_cancer()
```

## (3) n_neighbors 다양하게 설정

#### 데이터분할

```python
cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(
    cancer.data, cancer.target, stratify = cancer.target, random_state = 66)
```

#### k갯수 지정에 따른 훈련셋과 테스트셋 예측모델에 피팅

```python
# 최근접 이웃갯수 리스트객체 준비
neighbors_settings = range(1, 11)
list(neighbors_settings)
```

```python
# 모델피팅 결과저장용 리스트객체 준비
training_accuracy = []
test_accuracy = []
# - 최근접 이웃갯수에 따라 
#   훈련모델 피팅결과와 테스트결과의 정확도를 각각 저장할 공간임

# 파이썬에서 복수의 데이터셋을 갖을 수 있는 리스트객체, 딕셔너리 타입이 있는데
# 딕셔너리는 키값이 필요하므로 빈 리스트객체를 생성해 넣어줌
```

```python
# 최근접 이웃갯수에 따른 모델피팅
for k in neighbors_settings:
    # knn객체생성
    knn = KNeighborsClassifier(n_neighbors = k)
    
    # knn객체에 훈련데이터 피팅
    knn.fit(X_train, y_train)
    
    # 훈련데이터 피팅결과 정확도 저장
    training_accuracy.append(knn.score(X_train, y_train)) # 리스트명.append => 리스트 맨 뒤에 값 하나 추가
    
    # 테스트테이터와의 검증결과 정확도 저장
    test_accuracy.append(knn.score(X_test, y_test))
```

#### k갯수 지정에 따른 훈련셋과 테스트셋 정확도 성능비교

```python
import pandas as pd

training_accuracy_sr = pd.Series(training_accuracy)
test_accuracy_sr = pd.Series(test_accuracy)

model_perf = pd.DataFrame({'K지정 갯수':list(neighbors_settings),
                           '훈련셋 정확도': training_accuracy,
                           '테스트셋 정확도': test_accuracy,
                           '정확도 차이': test_accuracy_sr - training_accuracy_sr})
model_perf
```

```python
round(model_perf, 3)
```
